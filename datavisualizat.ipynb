```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', )
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.grid(True)
# Save the plot as a JPG file
plt.savefig("plot.jpg", format='jpg', dpi=300, bbox_inches='tight')
plt.show()
```
